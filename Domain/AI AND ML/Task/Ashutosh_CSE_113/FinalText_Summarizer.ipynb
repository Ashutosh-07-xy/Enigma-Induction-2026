```python
from transformers import pipeline

summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
```

```python
text = input("Enter the text to be summarized:\n")

if len(text.split()) < 30:
    print("\nText is too short to summarize. Please enter at least 30 words.")

else:
    print("\nChoose summary length:")
    print("  1 - Short \t 2 - Medium \t 3 - Long")
    choice = input("Enter choice (1/2/3): ").strip()

    if choice == "1":
        maxl, minl = 60, 20
    elif choice == "3":
        maxl, minl = 300, 100
    else:
        maxl, minl = 150, 50

    summary = summarizer(text, max_length=maxl, min_length=minl, do_sample=False)

    print("\n--- Summary ---")
    print(summary[0]["summary_text"])
```
